#If else condition

```python
age=int(input("Enter your age"))
if(age<=18):
  print("You are not adult")
else:
  print("You are adult")
```

Elif -Single input with multiple conditions

```python
age=int(input("Enter your age"))
if(age<18):
  print("You are adult")
elif(age>60):
    print("You are senior")
else:
      print("You are child")
```

For Loop -Multiple input with single condition

```python
lists=[12,45,78,94,65]
for age in lists:
  if(age<18):
   print("You are adult")
  elif(age>60):
    print("You are senior")
  else:
      print("You are child")
```
